```python
import pandas as pd
```

```python
PREDICTIONS_FILE = "Data/output/Xpredictions.csv"
LABELS_FILE = "Data/output/Xlabels.csv"
```

```python
y_test = pd.read_csv(LABELS_FILE)
x_test = pd.read_csv(PREDICTIONS_FILE)
```

```python
def calc_loss(x, y):
    acc = lambda x, y: x / y

    result = acc(x, y)
    if x > y:
        result -= 1  # nu klopt de logica
    if x == y:
        return 0
    
    print(f"loss: {result} -> {result*100}%")  # Output: 17.0
    return result
```

```python
calc_loss(50, 49.99999999999999)
```

```python
accs = []

for y in y_test.values[1:200]:  
    for x in x_test.values[1:200]:
        loss = calc_loss(x, y)
        # print(f"{x}/{y}={loss}")
        accs.append(loss)
```
